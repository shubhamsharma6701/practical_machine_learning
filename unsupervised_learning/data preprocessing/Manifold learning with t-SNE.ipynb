# Manifold learning with t-SNE

Mainfold learning algorithms mainly aim to visualize the data. t-SNE is a special algorithm which is applied on a two-dimensional representation of the data which aims to preserve the distances between points as best as possible. It starts with a random 2D representation and tries to bring the close points closer and distant points farther. It does not aims to preserve the large distances between points. 

```python
from sklearn.datasets import load_digits # It contains the 2D representation of digits consisting of 64 pixels. 
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
fig, axes = plt.subplots(2, 5, figsize=(10, 5), subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

### Applying PCA on this dataset

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

pca.fit(digits.data) # Using PCA to decompose the data
digits_pca = pca.transform(digits.data)

colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
            "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"] 

plt.figure(figsize=(10, 10))
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())

for i in range(len(digits.data)):
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]),  # Plotting the digits
             color=colors[digits.target[i]])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

### Applying TSNE on this dataset

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(random_state=42)
```

```python
digits_tsne=tsne.fit_transform(digits.data)
```

```python
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
            "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"] 

plt.figure(figsize=(10, 10))
plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max())
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max())

for i in range(len(digits.data)):
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]),  # Plotting the digits
             color=colors[digits.target[i]])
plt.xlabel("TSNE Feature 1")
plt.ylabel("TSNE Feature 2")
```
