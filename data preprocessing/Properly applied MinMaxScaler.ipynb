```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
x, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
```

```python
x_train, x_test = train_test_split(x, random_state=0, test_size=0.1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
```

```python
# Plotting the original data onto a graph 
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
axes[0].scatter(x_train[:, 0], x_train[:, 1], label="Training Data", s=20)
axes[0].scatter(x_test[:, 0], x_test[:, 1], label="Testing Data", s=20)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")
```

```python
# Plotting the badly scaled data in which the same value of minimum and maximum is used in scaling both training and testing 
# data

scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
axes[1].scatter(x_train_scaled[:, 0], x_train_scaled[:, 1], label="Training data", s=20)
axes[1].scatter(x_test_scaled[:, 0], x_test_scaled[:, 1], label="Testing data", s=20)
axes[1].legend(loc="upper left")
axes[1].set_title("Poorly scaled data")
```

```python
# Plotting the correctly scaled data in which the testing data is scaled independently of the training data and hence uses 
# the minimum and maximum of the testing set 

test_scaler = MinMaxScaler()
test_scaler.fit(x_test)
x_test_scaled_good = test_scaler.transform(x_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
axes[2].scatter(x_train_scaled[:, 0], x_train_scaled[:, 1], label="Training data", s=20)
axes[2].scatter(x_test_scaled_good[:, 0], x_test_scaled_good[:, 1], label="Testing data", s=20)
axes[2].legend(loc="upper left")
axes[2].set_title("Properly scaled data")
```
